```python
# Select only numeric variables
numeric_cols = college.select_dtypes(include=['int64', 'float64'])

# Calculate correlation matrix
corr_matrix = numeric_cols.corr()

corr_matrix
```

```python
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix,
            annot=True,
            cmap='coolwarm',
            linewidths=0.5,
            fmt='.2f')

plt.title("Correlation Heatmap of College Dataset", fontsize=16)
```

```python
# Unstack matrix and sort
corr_pairs = corr_matrix.unstack().sort_values(ascending=False)

# Show top 15 correlations (excluding 1.00 self-correlation)
corr_pairs[(corr_pairs < 0.999) & (corr_pairs > -0.999)].head(20)
```

```python
fig, axes = plt.subplots(1, 4, figsize=(22, 5))
fig.suptitle("Admission Pipeline with Private/Public Comparison", fontsize=16)

# 1. Applications vs Acceptances
sns.scatterplot(data=college, x="Apps", y="Accept", hue="Private", ax=axes[0], alpha=0.7)
sns.regplot(data=college, x="Apps", y="Accept", scatter=False, ax=axes[0], color="black")
axes[0].set_title("Applications vs Acceptances")
axes[0].set_xlabel("Applications")
axes[0].set_ylabel("Acceptances")

# 2. Acceptances vs Enrollments
sns.scatterplot(data=college, x="Accept", y="Enroll", hue="Private", ax=axes[1], alpha=0.7)
sns.regplot(data=college, x="Accept", y="Enroll", scatter=False, ax=axes[1], color="black")
axes[1].set_title("Acceptances vs Enrollments")
axes[1].set_xlabel("Acceptances")
axes[1].set_ylabel("Enrollments")

# 3. Enrollments vs Full-Time Undergrad Size
sns.scatterplot(data=college, x="F.Undergrad", y="Enroll", hue="Private", ax=axes[2], alpha=0.7)
sns.regplot(data=college, x="F.Undergrad", y="Enroll", scatter=False, ax=axes[2], color="black")
axes[2].set_title("Enrollments vs Full-Time Undergraduates")
axes[2].set_xlabel("F.Undergrad")
axes[2].set_ylabel("Enrollments")

# 4. Enrollments vs Part-Time Undergrad Size
sns.scatterplot(data=college, x="P.Undergrad", y="Enroll", hue="Private", ax=axes[3], alpha=0.7)
sns.regplot(data=college, x="P.Undergrad", y="Enroll", scatter=False, ax=axes[3], color="black")
axes[3].set_title("Enrollments vs Part-Time Undergraduates")
axes[3].set_xlabel("P.Undergrad")
axes[3].set_ylabel("Enrollments")
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

<u>What we can see:</u><br/>Public universities enroll significantly more students overall, accept more applicants, and include a broad mix of full-time and part-time students. Private universities are smaller and more selective, with higher proportions of full-time students and much lower part-time enrollment. This difference reflects the broader mission of public institutions to serve diverse and flexible student populations, while private universities tend to focus on full-time and smaller groups.

```python
#To take out the outliers showfliers=False
sns.boxplot(x='Private', y='Grad.Rate', data=college, showfliers=False)
```

```python
fig, axes = plt.subplots(2, 3, figsize=(18, 10))
fig.suptitle("Grade rate comparing with Private/Public", fontsize=16)

# 1. Top 10 percent student of H.S. class to grade rate
sns.scatterplot(data=college, x="Top10perc", y="Grad.Rate",
                hue="Private", ax=axes[0, 0], alpha=0.7)
sns.regplot(data=college, x="Top10perc", y="Grad.Rate",
            scatter=False, ax=axes[0, 0], color="black")
axes[0, 0].set_title("Top 10% of H.S. class vs Grade Rate")
axes[0, 0].set_xlabel("Top 10% of H.S. class")
axes[0, 0].set_ylabel("Grade Rate")

# 2. Out-of-state tuition compared to grade rate
sns.scatterplot(data=college, x="Outstate", y="Grad.Rate",
                hue="Private", ax=axes[0, 1], alpha=0.7)
sns.regplot(data=college, x="Outstate", y="Grad.Rate",
            scatter=False, ax=axes[0, 1], color="black")
axes[0, 1].set_title("Out-of-state tuition vs Grade Rate")
axes[0, 1].set_xlabel("Out-of-state tuition")
axes[0, 1].set_ylabel("Grade Rate")

# 3. Alumni who donate compared to grade rate
sns.scatterplot(data=college, x="perc.alumni", y="Grad.Rate",
                hue="Private", ax=axes[0, 2], alpha=0.7)
sns.regplot(data=college, x="perc.alumni", y="Grad.Rate",
            scatter=False, ax=axes[0, 2], color="black")
axes[0, 2].set_title("Alumni who donate vs Grade Rate")
axes[0, 2].set_xlabel("Alumni who donate (%)")
axes[0, 2].set_ylabel("Grade Rate")

# 4. Pct. of faculty with Ph.D.’s to grade rate
sns.scatterplot(data=college, x="PhD", y="Grad.Rate",
                hue="Private", ax=axes[1, 0], alpha=0.7)
sns.regplot(data=college, x="PhD", y="Grad.Rate",
            scatter=False, ax=axes[1, 0], color="black")
axes[1, 0].set_title("Faculty with PhD vs Grade Rate")
axes[1, 0].set_xlabel("Pct. of faculty with PhD")
axes[1, 0].set_ylabel("Grade Rate")

# 5. Pct. of faculty with terminal degree compared to grade rate
sns.scatterplot(data=college, x="Terminal", y="Grad.Rate",
                hue="Private", ax=axes[1, 1], alpha=0.7)
sns.regplot(data=college, x="Terminal", y="Grad.Rate",
            scatter=False, ax=axes[1, 1], color="black")
axes[1, 1].set_title("Faculty with terminal degree vs Grade Rate")
axes[1, 1].set_xlabel("Pct. of faculty with terminal degree")
axes[1, 1].set_ylabel("Grade Rate")

# 6. Instructional expenditure per student compared to grade rate
sns.scatterplot(data=college, x="Expend", y="Grad.Rate",
                hue="Private", ax=axes[1, 2], alpha=0.7)
sns.regplot(data=college, x="Expend", y="Grad.Rate",
            scatter=False, ax=axes[1, 2], color="black")
axes[1, 2].set_title("Expenditure per student vs Grade Rate")
axes[1, 2].set_xlabel("Instructional expenditure per student")
axes[1, 2].set_ylabel("Grade Rate")

plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

<u>Grade Rate Comapring </u> <br/>
Taken together, the six graphs show that private universities consistently achieve higher graduation rates than public universities. This advantage can be explained by several factors: they enroll more academically prepared students, charge higher tuition (reflecting greater institutional resources), receive stronger alumni support, and employ a higher proportion of highly qualified faculty. Private  universities also spend more per student, which directly correlates with stronger academic outcomes. It could be that student put in the effort because they have spent a lot of money to get their degree they want.

While public universities have strong faculty, they serve a broader and more diverse population with varying academic needs. The combination of student preparedness, institutional spending, and faculty quality helps explain why private universities on average, achieve higher graduation rates.

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
fig.suptitle("Cost comparising with Private/Public Comparison", fontsize=16)

# 1. Expend Instructional expenditure vs Top 10 percent student by Private/Public
sns.scatterplot(data=college, x="Expend", y="Top10perc", hue="Private", ax=axes[0], alpha=0.7)
sns.regplot(data=college, x="Expend", y="Top10perc", scatter=False, ax=axes[0], color="black")
axes[0].set_title("Expend  vs Top 10 percent student")
axes[0].set_xlabel("Instructional expenditure per student")
axes[0].set_ylabel("Top 10 percent student of H.S. class")

# 2. Expend Instructional expenditure vs Out-of-state tuition by Private/Public
sns.scatterplot(data=college, x="Expend", y="Outstate", hue="Private", ax=axes[1], alpha=0.7)
sns.regplot(data=college, x="Expend", y="Outstate", scatter=False, ax=axes[1], color="black")
axes[1].set_title("Expend vs Out-of-state")
axes[1].set_xlabel("Instructional expenditure per student")
axes[1].set_ylabel("Out-of-state tuition")

# 3. Out-of-state tuition vs Room and Board cost by Private/Public
sns.scatterplot(data=college, x="Outstate", y="Room.Board", hue="Private", ax=axes[2], alpha=0.7)
sns.regplot(data=college, x="Outstate", y="Room.Board", scatter=False, ax=axes[2], color="black")
axes[2].set_title("Out-of-state vs Room and Board cost")
axes[2].set_xlabel("Out-of-state tuition")
axes[2].set_ylabel("Room.Board Room and board costs")

plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

<u>Cost comparising</u> </br>
These cost-comparison plots show that private universities invest far more money per student and charge significantly higher tuition and living costs than public universities. This higher level of spending is closely associated with attracting academically strong students, as seen in the positive relationship between expenditure per student and the percentage of top-performing high school students. While it is possible that families with more financial resources are overrepresented at private universities, this cannot be concluded from the dataset alone, so it remains only a hypothesis.

Public universities, on the other hand, operate with lower tuition and lower per-student spending, which makes them accessible to a wider and more diverse population. Despite this lower per-student spending, public universities invest heavily in their academic programs and facilities, largely because they serve far more students overall. Their scale allows them to operate effectively even with lower tuition.

Overall, the patterns highlight substantial differences in financial resources and cost structures between private and public schools. Private operate with higher costs and higher spending per student, while public are more cost-efficient, serving larger populations at lower cost but with greater variation in student characteristics.
